### Daily Indicators:

- XUDLERS,Spot exchange rate - Euro into Sterling
- XUDLUSS,Spot exchange rate - US $ into Sterling

```python
daily_data = pd.read_csv('daily_rates.csv', skiprows=3, header=0)
```

```python
daily_data.head()
```

```python
daily_data['datetime'] = pd.to_datetime(daily_data['DATE'])
```

```python
plt.figure(figsize=(15,10))
plt.plot(daily_data['datetime'], daily_data['XUDLERS'], color='red', label='EUR/GBP')
plt.plot(daily_data['datetime'], daily_data['XUDLUSS'], color='blue', label='USD/GBP')
plt.legend(loc='best')
plt.xlabel('Year')
plt.ylabel('Euro and US Dollar to Pound exchange rate')
plt.grid(True)

plt.axvline(x=datetime.datetime(2016,1,2), color='green')
plt.axvline(x=datetime.datetime(2007,1,2), color='black')
plt.axvline(x=datetime.datetime(2009,1,12), color='black')
```

```python
plt.figure(figsize=(15,10))

brexit = daily_data['datetime'].apply(lambda row: row.year > 2015)
brexit_times = daily_data['datetime'][brexit]

plt.plot(brexit_times, daily_data['XUDLERS'][brexit], color='red', label='EUR/GBP')
plt.plot(brexit_times, daily_data['XUDLUSS'][brexit], color='blue', label='USD/GBP')
plt.legend(loc='best')
plt.xlabel('Year')
plt.ylabel('Euro and US Dollar to Pound exchange rate')
plt.grid(True)
```

```python
daily_data['dayofweek'] = daily_data['datetime'].apply(lambda row: row.dayofweek)
weekly_data = daily_data[daily_data['dayofweek'] == 4]
weekly_data.head()
```

```python
plt.figure(figsize=(15,10))

brexit = weekly_data['datetime'].apply(lambda row: row.year > 2015)
brexit_times = weekly_data['datetime'][brexit]

plt.plot(brexit_times, weekly_data['XUDLERS'][brexit], color='red', label='EUR/GBP')
plt.plot(brexit_times, weekly_data['XUDLUSS'][brexit], color='blue', label='USD/GBP')
plt.legend(loc='best')
plt.xlabel('Year')
plt.ylabel('Euro and US Dollar to Pound exchange rate')
plt.grid(True)
```

### Monthly Indicators:

- IUMAAJNB,Monthly average rate of discount - 3 month Treasury bills - Sterling
- IUMAJNB,End month level of discount rate - 3 month Treasury bills - Sterling
- IUMAVNEA,Monthly average Sterling 1 month mean interbank lending rate

- IUMVNEA,End month Sterling 1 month mean interbank lending rate 

- XUMAERS,Monthly average Spot exchange rate - Euro into Sterling
- XUMAUSS,Monthly average Spot exchange rate - US into Sterling
- XUMLERS,End month Spot exchange rate - Euro into Sterling
- XUMLUSS,End month Spot exchange rate - US into Sterling


```python
monthly_data = pd.read_csv('monthly_rates.csv', skiprows=11, header=0)
monthly_data.head()
```

```python
monthly_data['datetime'] = pd.to_datetime(monthly_data['DATE'])
```

```python
plt.figure(figsize=(15,10))
plt.plot(monthly_data['datetime'], monthly_data['XUMAUSS'], color='red', label='Monthly average Spot exchange rate - US into Sterling')
plt.plot(monthly_data['datetime'], monthly_data['XUMLUSS'], color='blue', label='End month Spot exchange rate - US into Sterling')
plt.legend(loc='best')
plt.xlabel('Year')
plt.ylabel('Euro and US Dollar to Pound exchange rate')
plt.grid(True)
```

```python
plt.figure(figsize=(15,10))

brexit = monthly_data['datetime'].apply(lambda row: row.year > 2015)
brexit_times = monthly_data['datetime'][brexit]

plt.plot(brexit_times, monthly_data['XUMAUSS'][brexit], color='red', label='Monthly average Spot exchange rate - US into Sterling')
plt.plot(brexit_times, monthly_data['XUMLUSS'][brexit], color='blue', label='End month Spot exchange rate - US into Sterling')
plt.legend(loc='best')
plt.xlabel('Year')
plt.ylabel('Euro and US Dollar to Pound exchange rate')
plt.grid(True)
```
