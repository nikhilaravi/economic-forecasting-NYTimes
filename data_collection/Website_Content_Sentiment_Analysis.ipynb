# SENTIMENT SECTION

#### SentiWordNet: Sentiment analysis of news articles was performed using SentiWordNet, a public dictionary of words with sentiment scores, which has been trained via semi-supervised learning. 

## SentiWordNet Methodology

### Selection of relevant articles
From all the articles scrapped using an API of The Gaurdian newspaper, we first filtered based on relevance to exchnage rates. This involve building a small dictionary of words and phrases which would indicate relevance to exchnage rate, and then search all the articles for the disctionary content. After some data exploration and visualisation, we determined the minimum number of words and phrases required for matching in order to class an article as being relevant (ensuring that a sufficient number of articles were classed as relevant weekly in order to generate weekly sentiment scores).

### - Tokenizing: article body text was split into sentences, and further into words
SentinetWordNet provides a trained index of words and their corresponding positive, negtiave and netural sentiments (where the sum of these sentiments equal 1). To generate sentiments at an article level, we tokenised articles into words and used the mean average sentiment as the sentiment score for the entire article. 

A potential problem behind this method is that we may not obtain accurate sentiments for sections of the artivle which depend on more sophisticated understand of the natural language at a sentence or a phrase level. However, given the computation time required to use more sophisticated natural language techniques (such as Word2Vec or Doc2Vec) and the total number of articles to be considered across the time period, we proceed with the simpler model.

### - Filtering Neutrality: Removing sentences with neutral sentiment from the original article
Given each article, we removed any sentences which had very high neutal sentiment scores. This ensured that our sentiment score generator would only focus on sentences which implied strong sentiment, resolving a previous problem of obtaining very low positive/negative sentiment scores from our model

### - Weekly sentiment: Averaging article sentiments at a weekly level to obtain weekly sentiment scores.
Having developed a method to generate sentiment scores for a given article, we simply took the mean average of the articles at a given week to generate the sentiment scores associated to that week. For future work, it may be worth considering potential ways to weight different articles' sentiments based on factors such as their popularity.

### - Smoothing: Using a rolling average of preceding weeks’ sentiments to reduce the weekly variance of sentiment scores.
Having generated weekly sentiment scores, initially the high weekly variance did not allow any visible trends to be noticed. We resolved this by smoothing the weekly sentiment scores using a rolling average of a fixed number of preceding weeks. The number of preceding weeks use to smoothen the scentiment scores was chosen by maximising the correlation between the exchange rate and the sentiment score time series. The smoothened sentiment scores now shows visible correlation with the exchange rate, especially during times such as the Brexit Referendum, as can be seen below.
